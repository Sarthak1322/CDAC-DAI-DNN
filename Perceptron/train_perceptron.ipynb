```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
from sklearn.datasets import make_classification

X,y = make_classification(n_samples=100, n_features=2, n_informative=1, n_redundant=0, n_clusters_per_class=1, 
                          n_classes=2, random_state=41, hypercube=False, class_sep=10)
# X is the data, y is the label

plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='winter', s=100)
```

```python
def perceptron(X,y):
    X = np.insert(X, 0, 1, axis=1) # add a column of ones to X
    weights = np.ones(X.shape[1]) # initialize weights to 1
    lr = 0.1

    for i in range(1000): # 1000 iterations
        j = np.random.randint(0, 100) # random index
        y_hat = step(np.dot(X[j], weights)) # predict
        weights = weights + lr * (y[j] - y_hat) * X[j] # update weights

    return weights[0],weights[1:]
```

```python
np.ones(3)
```

```python
def step(z):
    return 1 if z > 0 else 0 # step function
```

```python
intercept_ , coef_ = perceptron(X,y)
print(intercept_)
print(coef_)
```

```python
m = -(coef_[0] / coef_[1]) # slope
b = -(intercept_ / coef_[1]) # y intercept
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```
